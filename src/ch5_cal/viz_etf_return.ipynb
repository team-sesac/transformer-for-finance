```python
import matplotlib.pyplot as plt

#  
periods = ["1-month", "3-month", "6-month", "12-month"]
returns_without_ratio = [4.019320316514983, 2.138920875373258, 3.2469589725013743, 32.14348506021717]
returns_with_ratio = [4.23998868405425, 2.72598982492136, 4.780247918411473, 33.74103380789428]

#  
plt.figure(figsize=(10, 6))
plt.plot(periods, returns_without_ratio, label='Return without Ratio', marker='o')
plt.plot(periods, returns_with_ratio, label='Return with Ratio', marker='o')
plt.title('Portfolio Expected Returns Over Different Periods')
plt.xlabel('Period')
plt.ylabel('Expected Return (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#   
plt.subplot(3, 1, 2)
bar_width = 0.35
bar_positions_1 = range(len(returns_without_ratio))
bar_positions_2 = [x + bar_width for x in bar_positions_1]

plt.bar(bar_positions_1, returns_without_ratio, width=bar_width, label='Return without Ratio')
plt.bar(bar_positions_2, returns_with_ratio, width=bar_width, label='Return with Ratio')
plt.xticks([r + bar_width / 2 for r in range(len(periods))], periods)
plt.title('Bar Chart of Portfolio Expected Returns')
plt.xlabel('Period')
plt.ylabel('Expected Return (%)')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


file = 'passive_etf_stock_price_df.csv'
df = pd.read_csv(file, index_col=0).iloc[-252:]
ratio = pd.read_csv('../../portfolio/optimized_portfolio_ratio.csv', index_col=0)
returns_without_ratio = (df.iloc[-1] - df.iloc[0])/df.iloc[0]

returns_with_ratio = (df.iloc[-1] - df.iloc[0])/df.iloc[0] * ratio.iloc[:,0]

to_csv = pd.DataFrame([returns_without_ratio, returns_with_ratio], index=['returns_without_ratio', 'returns_with_ratio'], columns=df.columns)
to_csv.to_csv('passive_etf_returns_each_stock.csv')
```
